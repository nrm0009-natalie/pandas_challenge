```python
# Dependencies and Setup
import pandas as pd
```

```python
school_data_to_load = "/Users/admin/Documents/Data_Analytics_Class/pandas_challenge/Resources/schools_complete.csv"
student_data_to_load = "/Users/admin/Documents/Data_Analytics_Class/pandas_challenge/Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete
```

```python
#schools = school_data_complete["school_name"].value_counts()
total_schools = school_data_complete["school_name"].nunique()
total_schools
```

```python
total_students = school_data_complete["student_name"].count()
total_students
```

```python
#budget_df = school_data_complete["budget"].value_counts()
budget_df = school_data_complete["budget"].unique().sum()
budget_df
```

```python
avg_math_score = round(school_data_complete["math_score"].mean(), 6)
avg_math_score
```

```python
avg_read_score = round(school_data_complete["reading_score"].mean(), 6)
avg_read_score
```

```python
overall_passing_rate = ((avg_math_score + avg_read_score)/2)/100
overall_passing_rate
```

```python
number_passing_math = []
for item in school_data_complete["math_score"]:
    if item >= 70:
        number_passing_math.append(item)

len(number_passing_math)
```

```python
percent_passing_math = ((len(number_passing_math))/total_students)
percent_passing_math
```

```python
number_passing_reading = []
for item in school_data_complete["reading_score"]:
    if item >= 70:
        number_passing_reading.append(item)

len(number_passing_reading)
```

```python
percent_passing_reading = ((len(number_passing_reading))/total_students)
percent_passing_reading
```

```python
#making a dataframe and ordering columns correctly
district_summary = pd.DataFrame([{"Total Schools":total_schools,
                                  "Total Students":total_students,
                                  "Average Math Score":avg_math_score, 
                                 "Average Reading Score":avg_read_score, 
                                 "% Passing Math":percent_passing_math,
                                 "% Passing Reading":percent_passing_reading, 
                                 "% Overall Passing Rate":overall_passing_rate,
                                 "Total Budget":budget_df}])
organized_district_summary = district_summary[["Total Schools","Total Students","Total Budget",
                                               "Average Math Score",
                              "Average Reading Score","% Passing Math",
                              "% Passing Reading","% Overall Passing Rate"]]
organized_district_summary
```

```python
#Formatting rows
organized_district_summary["% Passing Math"] = organized_district_summary["% Passing Math"].map("{:.2%}".format)
organized_district_summary["% Passing Reading"] = organized_district_summary["% Passing Reading"].map("{:.2%}".format)
organized_district_summary["Total Budget"] = organized_district_summary["Total Budget"].map("${:.2f}".format)
#organized_district_summary["Total Budget"] = organized_district_summary["Total Budget"].map("{:,}".format)
organized_district_summary["% Overall Passing Rate"] = organized_district_summary["% Overall Passing Rate"].map("{:.2%}".format)
organized_district_summary
```

```python
# Calculate the total student count
per_school_counts = school_data_complete["school_name"].value_counts()
# Calculate the number of students that passed math in a dataframe
school_passing_math = school_data_complete[(school_data_complete["math_score"] >= 70)]

# Get the number of students that passed in a number
passing_math_count_total = school_passing_math.count()["student_name"]

#Get the Average passing math grade per school
passing_math_per_school = (school_passing_math.groupby(["school_name"]).count()["student_name"] / per_school_counts)*100

#change %passing to dataframe:
passing_math_df = pd.DataFrame(passing_math_per_school)
passing_math_df = passing_math_df.reset_index()
passing_math_df = passing_math_df.rename(columns={passing_math_df.columns[1]:"% Students Passing Math"})

passing_math_df

```

```python
per_school_counts = school_data_complete["school_name"].value_counts()

school_passing_reading = school_data_complete[(school_data_complete["reading_score"] >= 70)]
passing_reading_count_total = school_passing_reading.count()["student_name"]
passing_reading_per_school = (school_passing_reading.groupby(["school_name"]).count()["student_name"] / per_school_counts)*100


passing_reading_df = pd.DataFrame(passing_reading_per_school)
passing_reading_df = passing_reading_df.reset_index()
passing_reading_df = passing_reading_df.rename(columns={passing_reading_df.columns[1]:"% Students Passing Reading"})

passing_reading_df
```

```python
adding_columns = school_data_complete[["school_name","type","budget",
                                       "reading_score","math_score","size"]]

finding_averages = adding_columns.groupby(["school_name", "type"])
averages_by_school = finding_averages.mean()
averages_by_school["Per Student Budget"] = (averages_by_school["budget"])/(averages_by_school["size"])

averages_by_school = averages_by_school.rename(columns={"budget":"Total Budget",
                                                       "reading_score":"Average Reading Score",
                                                       "math_score":"Average Math Score",
                                                       "size":"Total Students"})


school_summary_df = pd.DataFrame(averages_by_school)
school_summary_df = school_summary_df.reset_index()
school_summary_df["% Students Passing Math"] = passing_math_df["% Students Passing Math"]
school_summary_df["% Students Passing Reading"] = passing_reading_df["% Students Passing Reading"]

#adding overall passing rate
school_summary_df["% Overall Passing Rate"] = (school_summary_df["% Students Passing Math"]+
school_summary_df["% Students Passing Reading"])/2
school_summary_df = school_summary_df.rename(columns=({"type":"Type"}))

#formatting
school_summary_formatting = school_summary_df
#school_summary_formatting["Students Passing Reading"] = school_summary_formatting["Students Passing Reading"].map("{:%}".format)
#school_summary_formatting["Students Passing Math"] = school_summary_formatting["Students Passing Math"].map("{:%}".format)
#school_summary_formatting["Overall Passing Rate"] = school_summary_formatting["Overall Passing Rate"].map("{:%}".format)
#school_summary_formatting["Total Budget"] = school_summary_formatting["Total Budget"].map("{:,}".format)
#school_summary_formatting["Per Student Budget"] = school_summary_formatting["Per Student Budget"].map("${:.0f}".format)

#reset index again back to school name
school_summary_formatting = school_summary_df.set_index("school_name")

school_summary_formatting
```
